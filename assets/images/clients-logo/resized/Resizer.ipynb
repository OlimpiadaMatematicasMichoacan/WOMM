```python
import fastai.vision.all as fv
import os
from pathlib import Path
import PIL
```

```python
src_path = Path(".")
dest_path = Path("resized")
```

```python
files = fv.get_files(src_path,recurse=False,extensions=['.png'])

dest_path.mkdir(exist_ok=True,parents=True)
```

```python
files
```

```python
def get_new_size(w,h, target_min_size = 512):
    factor = target_min_size/h
    return int(w*factor), int(h*factor)

def resizer(f):
    dest = dest_path/f"{f.relative_to(src_path).parent}"/f"{f.stem}.png"
    
    if dest.exists(): return # so that if something fails we don't have to restart from scratch
    
    dest.parent.mkdir(exist_ok=True, parents=True)
    
    img = PIL.Image.open(f)
    w,h = img.width, img.height
    
    resized = img.resize(get_new_size(w,h,target_min_size=200),resample=PIL.Image.LANCZOS)
    resized.save(dest)
```

```python
fv.parallel(resizer,files)
```
